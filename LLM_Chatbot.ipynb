# LLM Chatbot

In this project, we will build a ChatGPT-like chatbot powered by Qwen3.6

Import transformers and use pipeline to load the model and
import gradio to create interface

```python
!pip install -U gradio
from transformers import  pipeline # import pipeline from transformeres
import gradio as gr #Import Gradio and rename it as gr.
```

#Load the model

```python
#we will load the model into a variable named 'chatbot'
chatbot = pipeline(#pipeline is a high-level API that loads a model and tokenizer, and handles preprocessing, inference, and decoding.
    "text-generation",#This is the specific task type.
    model="Qwen/Qwen2.5-0.5B-Instruct",#the model name
    device_map="auto"#this controls where the model run (GPU or CPU)
)
```

###Create the function that will help us send a message to the chatbot and return the result.



```python
#Define the function with two parameters: message and  history.
def chat(message, history):
    messages = [  #create a list
        {"role": "system", "content": "You are a helpful AI assistant."},  #This is the system message; it tells the model how to behave.
    ]

    # add old conversation
    for user_msg, bot_msg in history:
        # add user message from history
        messages.append({"role": "user", "content": user_msg})

        # add bot response from history
        messages.append({"role": "assistant", "content": bot_msg})

    # add current user message
    messages.append({"role": "user", "content": message})

    #response will be the text generated by the model
    response = chatbot(  #calls the chatbot
        messages,  #sends the conversation message to the model
        max_new_tokens=200, #Limte the answer to maximum 200 new tokens.(Note: tokens can be a whole word or part of one )
        temperature=0.7,  #controls creativity.(0.1 :predictable, 0.7:balanced, 1.0+ more creative/random)
        do_sample=True #allows randomness in the answer
    )

    return response[0]["generated_text"][-1]["content"]#this extracts only the assistant's final answer.
```

###create interface using gradio

```python
css = """
.gradio-container {
    /* main app container */
    max-width: 900px !important;  /* limit width for clean layout */
    margin: auto !important;      /* center the app */
    font-family: Arial, sans-serif;  /* simple readable font */
}

#header {
    /* top title section */
    text-align: center;  /* center text */
    padding: 20px;       /* space inside header */
}

#header h1 {
    /* main title */
    font-size: 34px;     /* big title */
    margin-bottom: 5px;  /* space below title */
}

#header p {
    /* subtitle text */
    font-size: 16px;
    opacity: 0.8;        /* slightly faded look */
}

#chatbox {
    /* chat area */
    height: 550px;       /* fixed height */
    border-radius: 18px; /* rounded corners */
}

textarea {
    /* input box */
    border-radius: 14px !important;  /* rounded input */
}

button {
    /* all buttons */
    border-radius: 12px !important;  /* rounded buttons */
}
"""

with gr.Blocks(css=css, theme=gr.themes.Soft()) as demo:
    # create full UI app
    # apply CSS + theme

    gr.HTML("""
    <div id="header">
        <h1>🤖 My AI Chatbot</h1>
        <p>Clean chatbot interface with conversation memory</p>
    </div>
    """)
    # custom header (title + description)

    gr.ChatInterface(
        fn=chat,
        # function that runs when user sends message

        title=None,
        # remove default title (we use custom header)

        chatbot=gr.Chatbot(
            elem_id="chatbox"
            # link this to CSS #chatbox
        ),

        textbox=gr.Textbox(
            placeholder="Ask me anything...",
            # text before typing

            lines=2
            # input box height
        )
    )

demo.launch(debug=True)
# start app + show errors if exist
```
